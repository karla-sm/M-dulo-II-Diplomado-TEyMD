# $\color{purple}{\text{Variable aleatoria Cauchy}}$

Simulemos la variable aleatoria de Cauchy de parámetro 1 que se define de la siguiente forma la funcion de densidad de probabilidad:




$$ \mathbb{f_X}(x) = \frac{1}{\pi (1 + x^2)} \quad \text{si }  x \in\mathbb{R} $$

$\color{red}{\text{Ejercicio 1.}}$

***Corroborar que la función así definida es una función de densidad de probabilidad***

```python
import sympy as sp

# Declaramos x
x = sp.Symbol('x')

# Definimos la función de densidad de probabilidad
pdf = (1 / sp.pi) * (1 / (1 + x**2))

# Integramos la función de -∞ a ∞
integral = sp.integrate(pdf, (x, -sp.oo, sp.oo))

# Imprimimos el resultado
print(f"Resultado de la integral: {integral}")
```

El hecho de que obtengamos 1 como resultado de la integral nos demuestra que efectivamente es una función de densidad de probabilidad.

$\color{red}{\text{Ejercicio 2.}}$

 ***Encontrar La funcion de distribucion acumulariva*** $(\mathbb{F_X}(x))$

La función de distribución acumulativa $F_X(x)$ se define como:

$$F_X(x) = \int_{-\infty}^x f_X(t) \, dt$$

Usaremos la función de densidad que ya definimos:
$$f_X(t) = \frac{1}{\pi(1+t^2)}$$

```python
import sympy as sp

# Declaramos x y t
x, t = sp.symbols('x t')

# Definimos la función de densidad de probabilidad
pdf = (1 / sp.pi) * (1 / (1 + t**2))

# Calculamos la función de distribución acumulativa
cdf = sp.integrate(pdf, (t, -sp.oo, x))

# Imprimimos el resultado
print(f"F_X(x) = {cdf}")
```

Para que sea mejor entendible, lo que obtuvimos como resultado de $F_X(x)$  en el código, es lo mismo que:
$$F_X(x) = \frac{\arctan(x)}{\pi} + \frac{1}{2}$$


$\color{red}{\text{Ejercicio 3.}}$

***Crear una gráfica en Python que muestre su forma quiero el codigo en python***

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de densidad de la distribución de Cauchy
def cauchy_pdf(x):
    return 1 / (np.pi * (1 + x**2))

# Crear un rango de valores x
x = np.linspace(-10, 10, 1000)

# Calcular la densidad de probabilidad para cada valor de x
y = cauchy_pdf(x)

# Graficar la función de densidad
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribución de Cauchy", color="blue")
plt.title("Función de Densidad de Probabilidad de la Distribución de Cauchy")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

$\color{red}{\text{Ejercicio 4.}}$

***Generar V.A. Cauchy a partir de una uniforme***

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una variable aleatoria Cauchy
def generar_cauchy(n):
    U = np.random.uniform(0, 1, n)  # n valores uniformes
    X = np.tan(np.pi * (U - 0.5))  # Aplicar la fórmula
    return X

# Generar 1000 muestras de la distribución Cauchy
muestras_cauchy = generar_cauchy(1000)

# Filtrar valores extremos para centrarse en una parte más visible
muestras_centrales = muestras_cauchy[np.abs(muestras_cauchy) < 10]

# Crear un histograma de las muestras filtradas
plt.hist(muestras_centrales, bins=100, density=True, alpha=0.7, color='blue')

plt.title('Histograma Acercado de la Distribución Cauchy')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.show()
```
